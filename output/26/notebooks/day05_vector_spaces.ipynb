### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

import itertools as it

polynomial_samples = [
    np.array([a, b, c])
    for a, b, c in [(0, 0, 0), (1, 0, -1), (2, -1, 3), (-1, 4, 2)]
]


def closure_under_addition(vectors):
    for v1, v2 in it.product(vectors, repeat=2):
        result = v1 + v2
        if not any(np.allclose(result, candidate) for candidate in vectors):
            return False
    return True


def closure_under_scalars(vectors, scalars=(-2, -1, 0.5, 2)):
    for v in vectors:
        for s in scalars:
            result = s * v
            if not any(np.allclose(result, candidate) for candidate in vectors):
                return False
    return True

print('Closure under addition (sampled polynomials):', closure_under_addition(polynomial_samples))
print('Closure under scalar multiplication:', closure_under_scalars(polynomial_samples))

vectors = np.array([[0, 0], [2, 1], [1.5, 2.5]])
fig, ax = plt.subplots(figsize=(5, 5))
origin = np.array([0, 0])
for vec in vectors[1:]:
    ax.quiver(*origin, *vec, angles='xy', scale_units='xy', scale=1, color='#7570b3')
ax.quiver(*vectors[1], *(vectors[2] - vectors[1]), angles='xy', scale_units='xy', scale=1, color='#1b9e77', linestyle='--')
ax.set_xlim(-0.5, 3)
ax.set_ylim(-0.5, 3)
ax.set_aspect('equal')
ax.set_title('Reindeer Flight Parallelogram')
ax.grid(True, alpha=0.3)
plt.show()
```

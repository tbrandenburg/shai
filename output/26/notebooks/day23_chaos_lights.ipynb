### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

def logistic_map(r, x0, steps=100):
    xs = [x0]
    for _ in range(steps):
        xs.append(r * xs[-1] * (1 - xs[-1]))
    return xs

r_values = [2.8, 3.2, 3.5, 3.9]
plt.figure(figsize=(10, 4))
for r in r_values:
    xs = logistic_map(r, 0.2, 80)
    plt.plot(xs, label=f'r={r}')
plt.title('Light String Sensitivity')
plt.xlabel('Iteration')
plt.ylabel('Brightness')
plt.legend()
plt.show()

rs = np.linspace(2.5, 4.0, 400)
scatter_r = []
scatter_x = []
for r in rs:
    x = 0.4
    for _ in range(200):
        x = r * x * (1 - x)
    for _ in range(30):
        x = r * x * (1 - x)
        scatter_r.append(r)
        scatter_x.append(x)

plt.figure(figsize=(6, 6))
plt.scatter(scatter_r, scatter_x, s=1, color='#80b1d3')
plt.title('Bifurcation Sketch')
plt.xlabel('Parameter r')
plt.ylabel('Long-term brightness')
plt.show()
```

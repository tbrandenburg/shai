### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

ornament_matrix = np.array([
    [1, 2, 0, 4],
    [0, 1, 2, 3],
    [1, 3, 2, 7],
    [2, 4, 1, 9]
], dtype=float)

rank = np.linalg.matrix_rank(ornament_matrix)
print('Decor motif dimension:', rank)

matrix = sp.Matrix(ornament_matrix)
rref_matrix, pivots = matrix.rref()
print('Pivot columns (basis ornament styles):', pivots)

plt.figure(figsize=(6, 4))
plt.bar(range(matrix.shape[1]), np.linalg.norm(ornament_matrix, axis=0), color='#fdb462')
plt.xlabel('Vector index')
plt.ylabel('Norm')
plt.title('Relative ornament basis strengths')
plt.grid(True, axis='y', alpha=0.3)
plt.show()
```

### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

np.random.seed(0)
weights = [0.2, 0.35, 0.25, 0.2]
values = [1, 2, 4, 6]
data = np.random.choice(values, size=5000, p=weights)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(data, bins=np.arange(0.5, 6.5, 1), density=True, color='#fbb4ae')
plt.xlabel('Gift value')
plt.ylabel('Probability')
plt.title('Gift Lottery Distribution')

plt.subplot(1, 2, 2)
sorted_data = np.sort(data)
cdf = np.arange(1, len(data) + 1) / len(data)
plt.step(sorted_data, cdf)
plt.xlabel('Gift value')
plt.ylabel('CDF')
plt.title('Cumulative distribution')
plt.tight_layout()
plt.show()

print('Estimated expectation:', data.mean())
print('Estimated variance:', data.var())
```

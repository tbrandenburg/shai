### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

import cmath

def koch_segment(p, q, order):
    if order == 0:
        return [p]
    p = complex(*p)
    q = complex(*q)
    diff = q - p
    one_third = p + diff / 3
    two_third = p + diff * 2 / 3
    peak = one_third + diff / 3 * complex(-0.5, np.sqrt(3)/2)
    return (
        koch_segment((p.real, p.imag), (one_third.real, one_third.imag), order - 1)
        + koch_segment((one_third.real, one_third.imag), (peak.real, peak.imag), order - 1)
        + koch_segment((peak.real, peak.imag), (two_third.real, two_third.imag), order - 1)
        + koch_segment((two_third.real, two_third.imag), (q.real, q.imag), order - 1)
    )


def koch_snowflake(order):
    triangle = [(0, 0), (1, 0), (0.5, np.sqrt(3)/2), (0, 0)]
    curve = []
    for idx in range(len(triangle) - 1):
        curve.extend(koch_segment(triangle[idx], triangle[idx + 1], order))
    curve.append(triangle[-1])
    xs, ys = zip(*curve)
    return xs, ys


x_vals, y_vals = koch_snowflake(3)
plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals, color='#fa8072')
plt.axis('equal')
plt.title('Custom Fractal Ornament (Koch order 3)')
plt.axis('off')
plt.show()
```

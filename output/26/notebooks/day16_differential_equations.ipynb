### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

try:
    from scipy.integrate import solve_ivp
except Exception as exc:
    solve_ivp = None
    print('SciPy unavailable; slope-field only demonstration.', exc)

ambient = 5
initial_temp = 75
k = 0.15

def model(t, y):
    return -k * (y - ambient)

if solve_ivp:
    sol = solve_ivp(model, [0, 12], [initial_temp], t_eval=np.linspace(0, 12, 200))
    plt.figure(figsize=(6, 4))
    plt.plot(sol.t, sol.y[0], label='Numeric cooling')
    analytic = ambient + (initial_temp - ambient) * np.exp(-k * sol.t)
    plt.plot(sol.t, analytic, '--', label='Analytic model')
    plt.xlabel('Hours')
    plt.ylabel('Temperature (Â°C)')
    plt.title('Cocoa Cooling Curve')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    times = np.linspace(0, 12, 200)
    analytic = ambient + (initial_temp - ambient) * np.exp(-k * times)
    plt.plot(times, analytic)
    plt.show()
```

### Guided Experiments & Visuals
The following cell builds the math exploration plus the requested diagram.

```python

transformation = np.array([[2, -1], [1, 1]])
values, vectors = np.linalg.eig(transformation)
print('Eigenvalues:', values)

base_vectors = np.array([[1, 0], [0, 1], [1, 1], [-1, 2]])
transformed = base_vectors @ transformation.T

fig, ax = plt.subplots(figsize=(6, 6))
origin = np.zeros(2)
for vec, color in zip(base_vectors, ['#1b9e77', '#d95f02', '#7570b3', '#e7298a']):
    ax.quiver(*origin, *vec, color=color, angles='xy', scale_units='xy', scale=1, alpha=0.7)
for vec, color in zip(transformed, ['#1b9e77', '#d95f02', '#7570b3', '#e7298a']):
    ax.quiver(*origin, *vec, color=color, angles='xy', scale_units='xy', scale=1, linestyle='--')
for idx in range(vectors.shape[1]):
    eig_vec = vectors[:, idx].real
    ax.quiver(*origin, *(eig_vec * 3), color='#66a61e', width=0.007)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
ax.set_title('Gift Transformation Eigenvectors')
ax.grid(True, alpha=0.3)
plt.show()
```
